```python
df = df.replace('95 to 100%', 95).replace('70 to 75%', 70).replace('60 to 65%', 60).replace('35 to 40%', 35).replace('40 to 45%', 40)
df = df.replace('50 to 55%', 50).replace('55 to 60%', 55).replace('80 to 85%', 80).replace('25 to 30%', 25).replace('10 to 15%', 10)
df = df.replace('45 to 50%', 45).replace('65 to 70%', 65).replace('75 to 80%', 75).replace('85 to 90%', 85).replace('15 to 20%', 15)
df = df.replace('5% or less', 1).replace('5 to 10%', 5).replace('20 to 25%', 20).replace('90 to 95%', 90).replace('30 to 35%', 30).replace('50 to 65%', 50)
df.head()
```
